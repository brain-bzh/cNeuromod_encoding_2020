```python
Df = pd.read_csv('metrics_sub4.csv',index_col=0)
newmodelname = [f"{sub}_{ft}_{atlas}" for sub,ft,atlas in zip(Df['subject'],Df['finetune'],Df['atlas'])]
Df = Df.drop(columns=['model'])
Df.index = newmodelname
```
